## 1) Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall

#### Precision важна, когда в модели должно быть как можно больше положительно верных предсказаний. Нам не важны ложные предсказания в любом варианте. Поэтому нам не важно наполнение (recall).   Примером может послужить прогноз роста или падения цены на какой то товар. Чем чаще мы будет получать положительно верный ответ (то есть где будет цена в определенный момент времени), тем будет больше точность. Соответственно, придерживаясь этих рекомендаций биржевой игрок максимизирует прибыль, если будет придерживаться этих рекомендаций при торговле этим активом

#### Recall  важна когда важно соотношение наших любых предсказаний к количеству сделанных предсказаний. То есть мы должны оптимизировать таким образом модель, чтобы в расчет брались как положительные, так и негативные предсказания. В этом случае примером может служить прогноз на необходимость высаживания арбузов в средней полосе в зависимости от средней температуры мая за последние n-ное кол-во лет.  Мы смотрим сколько выросло арбузов и какова была P и R в прогнозах. Таким образом можно вычесть некую среднюю температуру мая, при которой имеет смысл высаживать арбузы предстоящим летом

## 2) Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

### Потому что F-метрика позволяет регулировать соотношение между P и R через коэффициент  β. Если мы его делаем больше 1, то увеличивается точность предсказаний за счет полноты (мы точнее, но реже это делаем). Если же мы делаем  β менее  1, то увеличиваем полноту, т.е. R
